### Leitura do arquivo

````python
import pandas as pd

df = pd.read_csv("../datasets/telemetria_nave_marte_simulada.csv")

# apenas para ver se deu certo a leitura
df.head()
````

### Algoritmo em python para interpretação dos dados

````python
import pandas as pd

# =========================
# LIMITES
# =========================
LIM_TEMP_INT_MIN = 18
LIM_TEMP_INT_MAX = 27

LIM_TEMP_EXT_MIN = -120
LIM_TEMP_EXT_MAX = 50

LIM_ENERGIA_MIN = 60
LIM_PRESSAO_MIN = 85

def identificar_falhas_linha(row):
    motivos = []
    if row["integridade_estrutural_bin"] != 1:
        motivos.append("Integridade")

    if row["energia_pct"] < LIM_ENERGIA_MIN:
        motivos.append("Energia baixa")

    if row["temperatura_interna_c"] < LIM_TEMP_INT_MIN or row["temperatura_interna_c"] > LIM_TEMP_INT_MAX:
        motivos.append("Temp_Interna")

    if row["temperatura_externa_c"] < LIM_TEMP_EXT_MIN or row["temperatura_externa_c"] > LIM_TEMP_EXT_MAX:
        motivos.append("Temp_Externa")

    if (
        row["pressao_tanque_combustivel_bar"] < LIM_PRESSAO_MIN or
        row["pressao_tanque_oxidante_bar"] < LIM_PRESSAO_MIN
    ):
        motivos.append("Pressão")

    if (
        row["status_propulsao"] != "OK" or
        row["status_navegacao"] != "OK" or
        row["status_comunicacao"] != "OK" or
        row["status_suporte_vida"] != "OK" or
        row["status_controle_termico"] != "OK"
    ):
        motivos.append("Módulos")

    return "|".join(motivos)

def verificar_decolagem():
  df["motivo_falha"] = df.apply(identificar_falhas_linha, axis=1)
  falhas = df[df["motivo_falha"] != ""]

  print(falhas[["timestamp_utc", "motivo_falha"]].to_string())

# CHAMA FUNÇÃO PARA PROCESSAR OS DADOS
verificar_decolagem()
````

# 1.4 Análise Energética
Calcular autonomia inicial considerando:
- Capacidade total (kwh);
- Carga atual (%);
- Consumo estimado na decolagem;
- Perdas energéticas.


## Solução
Pata a análise energética, foi utilizado o Python para interpretação dos dados

````python
# =========================
# CONSTANTES
# =========================
CAPACIDADE_TOTAL_KWH = 1200
CONSUMO_DECOLAGEM_KWH = 250
PERDAS_ENERGETICAS_PCT = 7

df["energia_disponivel_kwh"] = CAPACIDADE_TOTAL_KWH * (df["energia_pct"] / 100)
df["energia_util_kwh"] = df["energia_disponivel_kwh"] * (1 - PERDAS_ENERGETICAS_PCT / 100)
df["energia_restante_pos_decolagem_kwh"] = df["energia_util_kwh"] - CONSUMO_DECOLAGEM_KWH

df["status_autonomia"] = df["energia_restante_pos_decolagem_kwh"].apply(
    lambda x: "SUFICIENTE" if x > 0 else "INSUFICIENTE"
)

print(df[[
    "timestamp_utc",
    "energia_pct",
    "energia_disponivel_kwh",
    "energia_util_kwh",
    "energia_restante_pos_decolagem_kwh",
    "status_autonomia"
]].to_string())
````
